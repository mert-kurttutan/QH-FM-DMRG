Now, lets look at Particle Density Distribution throughout stages. This might indicate if there is an issue with the convergence

```python
#load the particle density to txt file, and then to numpy array, then plot as a function of x
chis = [100, 100, 200, 200, 400, 400, 400, 800, 800, 800, 800, 1600, 1600, 1600, 2000, 2000, 2000, 4000, 4000, 4000, 4000]
p2=ds.params()
p2.folder="QH_FM3"; p2.Lx=6; p2.Ly=4; p2.Nphi=4.; p2.U=8.0; p2.N=int(p2.Nphi); p2.S=0.; p2.pbc=True; 
p2.lat=ptn.mp.lat.su2u1.genFermiHubbardSpinCharge(p2.Lx*p2.Ly)
tar_loc="/home/m/Mert.Kurttutan/Academia/Codes/Physics/Projects/QH_Ferromagnetism/GS-Properties-2/dat-files-02-expc-arr/"

n_Arr = np.zeros((len(chis), p2.Lx))

for i in range(0,len(chis)):
    p2.bond=chis[i]; p2.ind=i
    print(i)
    try:
        mps_obj = ds.mps_load(p2)
        file_nm=ds.n_arr_save(mps_obj, tar_loc, p2)        #save it in the local dir
        n_arr = np.loadtxt(file_nm)
        n_Arr[i] = np.sum(n_arr, axis=1)

    except:
        print("State with " + "m_B=" + str(p2.bond) + " is not produced")
    
#ready for the plot    
```

```python
#plot
plt.figure(figsize=(8, 6), dpi=80)
# Using Numpy to create an array X
X = np.arange(0, p2.Lx, 1)

for i in range(3,len(chis)):
    plt.plot(X, n_Arr[i], '--o',label="m="+str(chis[i]))
    
    
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("$x$")
plt.ylabel("Particle Density - n(x)")
plt.title("Particle Density in x-direction")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend(loc ="upper right", prop={'size':8})
  
# To load the display window
plt.show()
    
```

Now, energy comparsion between $S_{max}$ and $S=0$ sector

```python
Lx_list = [6, 8, 12, 16]
Ly_list = [4, 5]
U = 8.0
s_arr = [0, 0.5]

E_arr = np.empty((len(Lx_list), len(Ly_list)), dtype=object)

loc_dir = "/home/m/Mert.Kurttutan/Academia/Codes/Physics/Projects/QH_Ferromagnetism/" + \
                "GS-Properties-2/dat-files-01-from-DMRG/"

for x_i in range(len(Lx_list)):
    for y_i in range(len(Ly_list)):
        Lx = Lx_list[x_i]
        Ly = Ly_list[y_i]
        N = 2
        E_arr[x_i, y_i] = np.zeros((Lx*Ly//4, 2))
        i=0
        while N <= Lx*Ly//2:
            Nphi = float(N)
            for s_i in range(len(s_arr)):
                
                S = s_arr[s_i]*Nphi
                fl_str = loc_dir + "Lx" + str(Lx) + "_Ly" + str(Ly) + "_Nphi" + str(Nphi) + "_U" + str(U) + "_N" + str(N) \
                            + "_S" + str(S) + "_variance_FHH_SU2.dat"

                # Using readlines()
                try:
                    file1 = open(fl_str, 'r')
                    Lines = file1.readlines()
                    file1.close()
                    #print(Lines[-1].split(","))
                
                    E = Lines[-1].split(",")[9]
                    E_arr[x_i, y_i][i, s_i] = E
                except:
                    print("nothing")
            i += 1
            N += 2 
            
```

```python
for x_i in range(len(Lx_list)):
    for y_i in range(len(Ly_list)):
        Lx=Lx_list[x_i]; Ly=Ly_list[y_i];
        l = np.shape(E_arr[x_i,y_i])[0]
        x_0 = 0.5/(l-1)
        X = np.linspace(x_0, 0.5, l)

        plt.figure(figsize=(8, 6))
        #plt.plot(X, E_arr[x_i,y_i][:,0], '-o', label='$S=0$')
        #plt.plot(X, E_arr[x_i,y_i][:,1], '-o', label='$S=S_{max}$')
        
        plt.plot(X, E_arr[x_i,y_i][:,0]- E_arr[x_i,y_i][:,1], '-o', label='diff  $E_{S=0} - E_{S=S_{max}}$ ')
        # Naming the x-axis, y-axis and the whole graph
        plt.xlabel("$Filling$")
        plt.ylabel("Ground State Energy Difference in units of t")
        plt.title("Comparison of GS Energy for $S=0$ and $S=S_{max}$ --- $L_x=$" + str(Lx) + " $ L_y=$"\
                  + str(Ly) + ", U=8")
        # Adding legend, which helps us recognize the curve according to it's color
        plt.legend(loc ="upper right", prop={'size':8})
        plt.grid()
        plt.savefig('dat-files-02-expc-arr/E_compare_02_' + str(Lx) + '-' + str(Ly) + '.png', dpi=100)
        plt.close()
print(E_arr[-1,0][:,0]- E_arr[-1,0][:,1])
```
